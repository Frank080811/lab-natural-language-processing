## Data Preprocessing

```python
# ------------------------------
# BASIC CLEANING FUNCTION (NO STOPWORDS)
# ------------------------------
import re

def clean_text_basic(text):
    text = str(text)

    # Remove JavaScript & CSS
    text = re.sub(r'<script.*?>.*?</script>', '', text, flags=re.DOTALL)
    text = re.sub(r'<style.*?>.*?</style>', '', text, flags=re.DOTALL)

    # Remove HTML comments
    text = re.sub(r'<!--.*?-->', '', text, flags=re.DOTALL)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # Remove numbers
    text = re.sub(r'\d+', ' ', text)

    # Remove single characters
    text = re.sub(r'\b[a-zA-Z]\b', ' ', text)

    # Remove single characters from start
    text = re.sub(r'^[a-zA-Z]\s+', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    # Lowercase
    text = text.lower().strip()

    return text

# Applying to train and test sets
X_train["text"] = X_train["text"].apply(clean_text_basic)
X_test["text"] = X_test["text"].apply(clean_text_basic)

print("âœ” Basic cleaning completed.")
```

## Now, we have to clean the html code removing words

- First we remove inline JavaScript/CSS
- Then we remove html comments. This has to be done before removing regular tags since comments can contain '>' characters
- Next we can remove the remaining tags

- Remove all the special characters
    
- Remove numbers
    
- Remove all single characters
 
- Remove single characters from the start

- Substitute multiple spaces with single space

- Remove prefixed 'b'

- Convert to Lowercase

```python
import nltk
from nltk.corpus import stopwords

# Download stopwords once
nltk.download("stopwords")

stop_words = set(stopwords.words("english"))

def remove_stopwords(text):
    words = text.split()
    clean_words = [w for w in words if w not in stop_words]
    return " ".join(clean_words)

# Apply to training and test text
X_train["text"] = X_train["text"].apply(remove_stopwords)
X_test["text"] = X_test["text"].apply(remove_stopwords)

print("âœ” Stopword removal completed.")
```

## Tame Your Text with Lemmatization
Break sentences into words, then use lemmatization to reduce them to their base form (e.g., "running" becomes "run"). See how this creates cleaner data for analysis!

```python
import nltk
from nltk.stem import WordNetLemmatizer

# Download resources (only once)
nltk.download("wordnet")
nltk.download("omw-1.4")

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = text.split()
    lem_words = [lemmatizer.lemmatize(w) for w in words]
    return " ".join(lem_words)

# Apply to training and test sets
X_train["text"] = X_train["text"].apply(lemmatize_text)
X_test["text"] = X_test["text"].apply(lemmatize_text)

print("âœ” Lemmatization completed.")
```

## Bag Of Words
Let's get the 10 top words in ham and spam messages (**EXPLORATORY DATA ANALYSIS**)

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Rebuild train dataset for analysis
train_df = pd.DataFrame({
    "text": X_train["text"],
    "label": y_train
})

# Separate ham (0) and spam (1)
ham_texts = train_df[train_df["label"] == 0]["text"]
spam_texts = train_df[train_df["label"] == 1]["text"]

# Create Bag of Words Vectorizer (remove English stopwords)
vectorizer = CountVectorizer(stop_words='english')

# Fit on HAM
ham_bow = vectorizer.fit_transform(ham_texts)
ham_word_counts = pd.DataFrame(ham_bow.toarray(), columns=vectorizer.get_feature_names_out())
top_ham = ham_word_counts.sum().sort_values(ascending=False).head(10)

# Fit on SPAM
vectorizer_spam = CountVectorizer(stop_words='english')
spam_bow = vectorizer_spam.fit_transform(spam_texts)
spam_word_counts = pd.DataFrame(spam_bow.toarray(), columns=vectorizer_spam.get_feature_names_out())
top_spam = spam_word_counts.sum().sort_values(ascending=False).head(10)

print("\nðŸ“Œ TOP 10 WORDS IN HAM MESSAGES:")
print(top_ham)

print("\nðŸ“Œ TOP 10 WORDS IN SPAM MESSAGES:")
print(top_spam)
```

```python
import seaborn as sns


plt.style.use("seaborn-v0_8-darkgrid")
sns.set_palette("viridis")

plt.figure(figsize=(16, 6))

def plot_top_words(data, title, color):
    # Create a dummy hue
    hue_values = list(range(len(data)))

    ax = sns.barplot(
        x=data.index,
        y=data.values,
        hue=hue_values,           # REQUIRED for future versions
        dodge=False,              # Ensures bars stay single
        palette=sns.color_palette(color, len(data)),
        legend=False              # Hide the hue legend
    )

    plt.title(title, fontsize=16, fontweight='bold')
    plt.xticks(rotation=45, ha='right', fontsize=12)

    # Add bar labels
    for i, val in enumerate(data.values):
        ax.text(i, val + max(data.values)*0.02, str(val),
                ha='center', fontsize=10, color='white', fontweight='bold')

    sns.despine()

# ===========================
plt.subplot(1, 2, 1)
plot_top_words(top_ham, "Top 10 Words in HAM Messages", "Blues")

plt.subplot(1, 2, 2)
plot_top_words(top_spam, "Top 10 Words in SPAM Messages", "Reds")

plt.tight_layout()
plt.show()
```

# Extra Features

```python
def add_advanced_features(df):
    text_col = "text"

    money_simbol_list = "|".join(["euro","dollar","pound","â‚¬", r"\$"])

    suspicious_words = "|".join([
        "free","cheap","sex","money","account","bank",
        "fund","transfer","transaction","win","deposit","password"
    ])

    # Non-capturing phone number regex
    phone_pattern = r"(?:\+?\d{1,3}?[-.\s]??\d{2,4}?[-.\s]??\d{3,4}?[-.\s]??\d{3,4})"

    df["money_mark"] = df[text_col].str.contains(money_simbol_list, regex=True).astype(int)
    df["suspicious_words"] = df[text_col].str.contains(suspicious_words, regex=True).astype(int)
    
    df["text_len"] = df[text_col].apply(len)
    df["exclamation_count"] = df[text_col].str.count("!")

    df["uppercase_ratio"] = df[text_col].apply(
        lambda x: sum(1 for c in x if c.isupper()) / max(1, len(x))
    )

    df["has_phone_number"] = df[text_col].str.contains(phone_pattern, regex=True).astype(int)

    return df
```

```python
# Apply feature engineering AFTER stopword removal
train_df = pd.DataFrame({
    "text": X_train["text"],
    "label": y_train
})

val_df = pd.DataFrame({
    "text": X_test["text"],
    "label": y_test
})

# Add advanced NLP features
train_df = add_advanced_features(train_df)
val_df   = add_advanced_features(val_df)

print("âœ… Advanced NLP features successfully added!")
train_df.head()
```

## How would work the Bag of Words with Count Vectorizer concept?

```python
# ============================================
# BAG OF WORDS USING COUNTVECTORIZER
# ============================================

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from scipy.sparse import hstack

# --------------------------------------------
# Ensure text column is clean
# --------------------------------------------
train_df["text"] = train_df["text"].astype(str)

# --------------------------------------------
# Initialize CountVectorizer (BOW)
# --------------------------------------------
vectorizer = CountVectorizer(
    max_features=3000,       # choose top 3000 frequent words
    ngram_range=(1, 2),      # include unigrams + bigrams
    stop_words="english"     # remove English stopwords
)

# --------------------------------------------
# Fit the vectorizer on the training text
# --------------------------------------------
X_bow = vectorizer.fit_transform(train_df["text"])

print("Shape of Bag-of-Words matrix:", X_bow.shape)
print("Example Vocabulary:", list(vectorizer.vocabulary_.items())[:10])

# --------------------------------------------
# Selecting your engineered features
# --------------------------------------------
extra_features = train_df[[
    "money_mark",
    "suspicious_words",
    "text_len",
    "exclamation_count",
    "uppercase_ratio",
    "has_phone_number"
]]

# Convert to numeric matrix
X_extra = extra_features.values

print("Shape of manually engineered features:", X_extra.shape)

# --------------------------------------------
# Combining BOW + engineered features
# --------------------------------------------
X_final = hstack([X_bow, X_extra])

print("Final combined feature matrix:", X_final.shape)

# --------------------------------------------
# Target variable
# --------------------------------------------
y_final = train_df["label"].values

# --------------------------------------------
# Train-test split
# --------------------------------------------
X_train_bow, X_test_bow, y_train_bow, y_test_bow = train_test_split(
    X_final, y_final, test_size=0.2, random_state=42, stratify=y_final
)

print("Train shape:", X_train_bow.shape)
print("Test shape:", X_test_bow.shape)
```

## TF-IDF

- Load the vectorizer

- Vectorize all dataset

- print the shape of the vetorized dataset

```python
# ============================================
# TF-IDF VECTORIZATION
# ============================================

from sklearn.feature_extraction.text import TfidfVectorizer

# --------------------------------------------
# Ensure text column is clean and string type
# --------------------------------------------
train_df["text"] = train_df["text"].astype(str)

# --------------------------------------------
# Initialize the TF-IDF Vectorizer
# --------------------------------------------
tfidf = TfidfVectorizer(
    max_features=5000,        # limit vocabulary size
    ngram_range=(1, 2),       # unigrams + bigrams
    stop_words="english",     # remove English stopwords
    min_df=2,                 # ignore rare words
    max_df=0.95               # ignore extremely common words
)

# --------------------------------------------
# Fit the vectorizer on ALL text & transform
# --------------------------------------------
X_tfidf = tfidf.fit_transform(train_df["text"])

# --------------------------------------------
# Print results
# --------------------------------------------
print("\n===== TF-IDF Results =====")
print("TF-IDF matrix shape:", X_tfidf.shape)
print("Vocabulary size:", len(tfidf.vocabulary_))
print("Example vocabulary items:", list(tfidf.vocabulary_.items())[:10])
```

## And the Train a Classifier?

```python
# ============================================
# TRAINING A MODEL (Linear SVM)
# ============================================

from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# --------------------------------------------
# Target variable
# --------------------------------------------
y = train_df["label"]

# --------------------------------------------
# Train-test split 
# --------------------------------------------
X_train_tfidf, X_test_tfidf, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.20, random_state=42, stratify=y
)

print("Train shape:", X_train_tfidf.shape)
print("Test shape:", X_test_tfidf.shape)

# --------------------------------------------
# TRAIN THE  MODEL FOR THIS TASK
# --------------------------------------------
model = LinearSVC()

print("\nTraining Linear SVM model...")
model.fit(X_train_tfidf, y_train)
print("âœ” Model training completed successfully!")

# --------------------------------------------
# PREDICTION
# --------------------------------------------
y_pred = model.predict(X_test_tfidf)

```

```python
# --------------------------------------------
# EVALUATION
# --------------------------------------------
print("\nðŸ“Œ Accuracy:", round(accuracy_score(y_test, y_pred), 4))
print("\nðŸ“Œ Classification Report:\n")
print(classification_report(y_test, y_pred))

```

```python
# --------------------------------------------
# CONFUSION MATRIX VISUALIZED
# --------------------------------------------
plt.figure(figsize=(6,5))
sns.heatmap(
    confusion_matrix(y_test, y_pred),
    annot=True,
    fmt='d',
    cmap='Blues'
)
plt.title("Confusion Matrix â€” Linear SVM (TF-IDF)", fontsize=16)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
